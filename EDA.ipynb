```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stud_data = pd.read_csv("raw/online_course_engagement_data.csv")
stud_data.head(5)
courses=stud_data['CourseCategory'].unique()
courses
```

```python
stud_data.hist(bins=30, figsize=(15,10))
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(data=stud_data,x='CourseCategory')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=stud_data, x='DeviceType')
plt.show()
```

```python
numeric_df = stud_data.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(5,3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(data=stud_data,x='CourseCompletion',y='TimeSpentOnCourse')
plt.title('Time Spent on Course by Course Completion')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(data=stud_data ,x="CourseCompletion", y='QuizScores')
plt.title("Quiz Scores by Course Completion")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data=stud_data,x='DeviceType', hue='CourseCompletion')
plt.title('Course Completion by Device Type')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=stud_data, x='CourseCategory', hue='CourseCompletion')
plt.title('Course Completion by Course Category')
plt.xticks(rotation=45)
plt.show()
```
